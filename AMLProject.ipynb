### Creating the Transition Matrices

```python
lex = {'DDD':0,'DDL':1,'DDW':2,'DLD':3,'DLL':4,'DLW':5,'DWD':6,'DWL':7,'DWW':8,'LDD':9,'LDL':10,'LDW':11,'LLD':12,
      'LLL':13,'LLW':14,'LWD':15,'LWL':16,'LWW':17,'WDD':18,'WDL':19,'WDW':20,'WLD':21,'WLL':22,'WLW':23,'WWD':24,
      'WWL':25,'WWW':26}
```

```python
dmat = {}
for i in range(1,21):
    dmat[i] = np.zeros((27,27), dtype = int)
for i in range(9,19):
    print("Current Dataset: {}".format(i))
    for j in range(1,21):
        print("Current Team: {}".format(j))
        l = []
        for index,row in d[i].iterrows():
            if row['HTN'] == j or row['ATN'] == j:
                if row['HTN'] == j:
                    if row['FTR'] == 'D':
                        l.append('D')
                    elif row['FTHG'] > row['FTAG']:
                        l.append('W')
                    else:
                        l.append('L')
                else:
                    if row['FTR'] == 'D':
                        l.append('D')
                    elif row['FTAG'] > row['FTHG']:
                        l.append('W')
                    else:
                        l.append('L')
            if len(l)>3:
                dmat[j][lex[''.join(l[len(l)-4:len(l)-1])]][lex[''.join(l[-3:])]] += 1
```

```python
trans = {}
for i in range(1,21):
    trans[i] = dmat[i]/dmat[i].sum(axis=1)[:,None]
```

### Training Data Generator

```python
def pastgen(df, opp, index, verbose = 0):
    li = []
    sti = []
    gsi = []
    for i,r in df.iterrows():
        if verbose == 1:
            print('Fixture {}'.format(i))
            print('li is {}'.format(li))
            print('gdi is {}'.format(gdi))
        if i >= index:
            return [li[-3:],sum(gsi[-3:]),sum(sti[-3:])]
        if r['HomeTeam'] == opp or r['AwayTeam'] == opp:
            if r['HomeTeam'] == opp:
                sti.append(row['HST'])
                gsi.append(r['FTHG'])
                if r['FTR'] == 'D':
                      li.append('D')
                elif r['FTHG'] > r['FTAG']:
                    li.append('W')
                else:
                    li.append('L')
            else:
                sti.append(row['AST'])
                gsi.append(r['FTAG'])
                if r['FTR'] == 'D':
                    li.append('D')
                elif r['FTAG'] > r['FTHG']:
                    li.append('W')
                else:
                    li.append('L')
```

```python
priordict = {}
for i in range(1,21):
    priordict[i] = [0,0,0]
for i in range(9,14):
    for index, row in d[i].iterrows():
        if row['FTR'] == 'D':
            priordict[row['HTN']] = [sum(x) for x in zip(priordict[row['HTN']],[1,0,0])]
            priordict[row['ATN']] = [sum(x) for x in zip(priordict[row['ATN']],[1,0,0])]
        elif row['FTR'] == 'H':
            priordict[row['HTN']] = [sum(x) for x in zip(priordict[row['HTN']],[0,0,1])]
            priordict[row['ATN']] = [sum(x) for x in zip(priordict[row['ATN']],[0,1,0])]
        else:
            priordict[row['HTN']] = [sum(x) for x in zip(priordict[row['HTN']],[0,1,0])]
            priordict[row['ATN']] = [sum(x) for x in zip(priordict[row['ATN']],[0,0,1])]
for i in range(1,21):
    priordict[i] = [x / sum(priordict[i]) for x in priordict[i]]
```
